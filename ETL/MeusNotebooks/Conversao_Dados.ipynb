##### Importando as bibliotecas

```python
import pandas as pd
```

##### Carregando o arquivo

```python
Caminho = r'C:\Python\ETL\ArquivosCSV\Clientes.csv'

df = pd.read_csv(Caminho, sep=',', encoding='utf-8')

df
```

#### Listando os tipos de dados

```python
df.dtypes
```

##### Alterando o tipo de dado de algumas colunas

```python
# Dicionário com os tipos desejados
tipos = {
    'created_at': 'datetime64[ns]',
    'number': 'str',
}

# Aplicando a conversão
df = df.astype(tipos)

df.dtypes
```

```python
df
```

##### Tratando os dados missing

```python
# Criando uma cópia do DataFrame
df_silver = df.copy()

# Tratando colunas de texto
colunas_texto = ['email', 'state', 'street', 'additionals']
df_silver[colunas_texto] = df_silver[colunas_texto].fillna('Nao Informado')

# Mostrando o DataFrame tratado
df_silver
```

##### Filtrando dados pelo campo de data

```python
# Verificando qual é a maior e menor data que possuo no Dataframe

menor_data = df_silver['created_at'].min()
maior_data = df_silver['created_at'].max()
print(f'Menor data: {menor_data}')
print(f'Maior data: {maior_data}')
```

```python
# Filtro para pegar os dados entre duas datas

DataInicial = '2017-10-01'
DataFinal = '2017-12-31'

df_L6M = df_silver[(df_silver['created_at'] >= DataInicial) & (df_silver['created_at'] <= DataFinal)]

df_L6M
```

```python
# Filtrando apenas um ano específico
df_2017 = df_silver[df_silver['created_at'].dt.year == 2017]

df_2017
```

```python
# Filtrando dados por um ano e mes específico
df_2017_10 = df_silver[(df_silver['created_at'].dt.year == 2017) & (df_silver['created_at'].dt.month == 10)]
df_2017_10
```

##### Inserindo colunas no Dataframe

```python
# Inserindo colunas no Dataframe
df_silver['year'] = df_silver['created_at'].dt.year
df_silver['month'] = df_silver['created_at'].dt.month
df_silver['day'] = df_silver['created_at'].dt.day
df_silver['FullName'] = df_silver['first_name'] + ' ' + df_silver['last_name']

df_silver   
```

```python
# inserindo uma coluna com o nome do mes abreviado usando .map e a função lambda
Months = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'Mai', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

df_silver['MonthName'] = df_silver['created_at'].dt.month.map(Months)

df_silver
```
