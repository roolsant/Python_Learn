##### Importando as bibliotecas

```python
import pandas as pd
```

##### Carregando o arquivo

```python
Caminho = r'C:\Python\ETL\ArquivosCSV\Clientes.csv'

df = pd.read_csv(Caminho, sep=',', encoding='utf-8')

df
```

#### Listando os tipos de dados

```python
df.dtypes
```

##### Alterando o tipo de dado de algumas colunas

```python
# Dicionário com os tipos desejados
tipos = {
    'created_at': 'datetime64[ns]',
    'number': 'str',
}

# Aplicando a conversão
df = df.astype(tipos)

df.dtypes
```

```python
df
```

##### Tratando os dados missing

```python
# Criando uma cópia do DataFrame
df_silver = df.copy()

# Tratando colunas de texto
colunas_texto = ['email', 'state', 'street', 'additionals']
df_silver[colunas_texto] = df_silver[colunas_texto].fillna('Nao Informado')

# Mostrando o DataFrame tratado
df_silver
```
