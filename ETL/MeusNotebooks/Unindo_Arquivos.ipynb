##### Importando as bibliotecas

```python
import pandas as pd
import glob
import os
```

##### Carregando os arquivos da pasta

```python
# Caminho da pasta com os arquivos
caminho = r'C:\Python\ETL\ArquivosExcel'

# Pega todos os arquivos que começam com "Vendas" e terminam com .xlsx
arquivos = glob.glob(caminho + r"\Vendas*.xlsx")

# Lê e concatena todos os arquivos em um único DataFrame
df_vendas = pd.concat(
    [pd.read_excel(arquivo) for arquivo in arquivos], 
    ignore_index=True)

# Visualiza os primeiros registros
df_vendas
```

##### Salvando os arquivos consolidados em Excel

```python
# Caminho da pasta de destino
Pasta_consolidada = r'C:\Python\ETL\ArquivosExcel\Vendas_Consolidado'

# Salva o DataFrame em Excel
df_vendas.to_excel(os.path.join(Pasta_consolidada, 'Vendas_Consolidadas.xlsx'), index=False)

# Mensagem de confirmação
print('Arquivo salvo na pasta:', Pasta_consolidada)
```

##### Salvando os arquivos consolidados em CSV

```python
# Caminho da pasta de destino
Pasta_consolidada = r'C:\Python\ETL\ArquivosExcel\Vendas_Consolidado'

# Caminho completo do arquivo CSV
arquivo_csv = os.path.join(Pasta_consolidada, 'Vendas_Consolidadas.csv')

# Salva o DataFrame em CSV
df_vendas.to_csv(arquivo_csv, index=False, encoding='utf-8-sig')

# Mensagem de confirmação
print(f'Arquivo CSV salvo com sucesso em: {arquivo_csv}')
```

##### Salvando os arquivos consolidados em TXT

```python
# Caminho da pasta de destino
Pasta_consolidada = r'C:\Python\ETL\ArquivosExcel\Vendas_Consolidado'

# Caminho completo do arquivo TXT
arquivo_txt = os.path.join(Pasta_consolidada, 'Vendas_Consolidadas.txt')

# Salva o DataFrame em TXT
df_vendas.to_csv(arquivo_txt, index=False, sep=',', encoding='utf-8-sig')

# Mensagem de confirmação
print(f'Arquivo TXT salvo com sucesso em: {arquivo_txt}')
```
