##### Importando as bibliotecas

```python
import pandas as pd
```

##### Criando o Dataframe

```python
Caminho_Arquivo = r'C:\Python\ETL\ArquivosCSV\Clientes.csv'

df = pd.read_csv(Caminho_Arquivo, sep=',')
df

```

##### Listar valores distintos de uma coluna

```python
Coluna = 'state'

dfDistintos = df[Coluna].unique()
dfDistintos
```

##### Criando um novo Dataframe com o estados selecionados
###### *Se forem necessárias modificações futuras nesse novo df o ideal é usar o .copy() para evitar o erro SettingWithCopyWarning*

```python
Reg_Sul = ['Paraná', 'Santa Catarina','Rio Grande do Sul']

df_Sul = df.loc[df[Coluna].isin(Reg_Sul)].copy()
df_Sul
```

##### Tratando dados Missing

```python
### Identificando valores Missing por coluna
Missing = df.isnull().sum()
Missing = Missing[Missing > 0]
Missing
```

##### Iniciando tratativas nos dados
###### *usar o .copy() para criar um novo Dataframe e evitar o erro SettingWithCopyWarning*

```python
# Criando um novo dataframe para iniciar tratativas nos dados

df_silver = df.copy()
df_silver
```

##### Tratando os dados missing

```python
# Tratando as colunas de texto com valores missing

cols_to_fill = ['email', 'state', 'street', 'number', 'additionals']
df_silver[cols_to_fill] = df_silver[cols_to_fill].fillna('Nao Informado')


df_silver
```

##### Usando o método .replace

```python
# Dicionário: nome do estado -> sigla
mapeamento_estados = {
    'Acre': 'AC',
    'Alagoas': 'AL',
    'Amapá': 'AP',
    'Amazonas': 'AM',
    'Bahia': 'BA',
    'Ceará': 'CE',
    'Distrito Federal': 'DF',
    'Espírito Santo': 'ES',
    'Goiás': 'GO',
    'Maranhão': 'MA',
    'Mato Grosso': 'MT',
    'Mato Grosso do Sul': 'MS',
    'Minas Gerais': 'MG',
    'Pará': 'PA',
    'Paraíba': 'PB',
    'Paraná': 'PR',
    'Pernambuco': 'PE',
    'Piauí': 'PI',
    'Rio de Janeiro': 'RJ',
    'Rio Grande do Norte': 'RN',
    'Rio Grande do Sul': 'RS',
    'Rondônia': 'RO',
    'Roraima': 'RR',
    'Santa Catarina': 'SC',
    'São Paulo': 'SP',
    'Sergipe': 'SE',
    'Tocantins': 'TO'
}

# Substituindo na coluna 'state'
df_silver['state'] = df_silver['state'].replace(mapeamento_estados)

df_silver
```
